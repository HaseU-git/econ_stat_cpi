## データの準備

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fm.findSystemFonts()

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df_us_2019 = pd.read_csv("us_cpi_2019.csv", index_col=0).drop(["Relative\nimportance\nSep.\n2019", "2019-10-01"], axis=1)
```

```python
df_us_2019.head()
```

```python
df_us_2020 = pd.read_csv("us_cpi_2020.csv", index_col=1).drop(["Indent Level", "Relative\nimportance\nSep.\n2020", "2020-10-01"], axis=1)
```

```python
df_us_2020.head()
```

```python
df_us_2021 = pd.read_csv("us_cpi_2021.csv", index_col=1).drop(["Indent Level", "Relative\nimportance\nSep.\n2021"], axis=1)
```

```python
df_us_2021.head()
```

```python
df_us_all = pd.merge(df_us_2019, df_us_2020, on='Expenditure category', how='inner')
df_us_all = pd.merge(df_us_all, df_us_2021, on='Expenditure category', how='inner')
df_us_all.head()
```

```python
jp_df = pd.read_csv("jp_cpi.csv", encoding="SHIFT-JIS", index_col=0, header=1).T
jp_df = jp_df.drop(jp_df.columns[0:3], axis=1)
jp_df["weight"] = jp_df["ウエイト１万分比(Weight per 10000)"].astype("float") / 100
jp_df = jp_df.drop(["ウエイト１万分比(Weight per 10000)"], axis=1)
jp_df = jp_df.astype("float64")
```

```python
jp_df.head()
```

```python
jp_df_detail = pd.read_csv("jp_cpi_detail.csv", encoding="SHIFT-JIS", index_col=0, header=1).T
jp_df_detail = jp_df_detail.drop(jp_df.columns[0:3], axis=1)
jp_df_detail["weight"] = jp_df_detail["ウエイト１万分比(Weight per 10000)"].astype("float") / 100
jp_df_detail = jp_df_detail.drop(["ウエイト１万分比(Weight per 10000)"], axis=1)
jp_df_detail = jp_df_detail.astype("float64")
```

```python
jp_df_detail.head()
```

## 日本のデータの分析

```python
jp_rate_df = pd.DataFrame()
i = jp_df.columns.get_loc("201810")
index = i
month = 25

while i - index < month:
    tmp = (jp_df[jp_df.columns[i + 12]] - jp_df[jp_df.columns[i]]) / jp_df[jp_df.columns[i]] * 100
    jp_rate_df = pd.concat([jp_rate_df, tmp], axis=1)
    i = i + 1
jp_rate_df.columns = jp_df.columns[index + 12 : index + 12 + month]
```

```python
jp_rate_df = pd.DataFrame()
i = jp_df_detail.columns.get_loc("201810")
index = i
month = 25

while i - index < month:
    tmp = (jp_df_detail[jp_df_detail.columns[i + 12]] - jp_df_detail[jp_df_detail.columns[i]]) / jp_df_detail[jp_df_detail.columns[i]] * 100
    jp_rate_df = pd.concat([jp_rate_df, tmp], axis=1)
    i = i + 1
jp_rate_df.columns = jp_df_detail.columns[index + 12 : index + 12 + month]
```

```python
jp_rate_df.sort_values("202110", ascending=False).head(10)
# us_rate_df.head()
```

```python
jp_rate_df.sort_values("202110", ascending=False).T.plot.line(figsize=(20, 20))
plt.legend(loc='upper left', fontsize=10)
# plt.savefig("image_jp.png")
```
